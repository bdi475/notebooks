    - `avg_price`
- ‚úîÔ∏è The columns of `df_avg_and_count_by_brand` should be a flat-level index, not a multi-level index.
    - **To check:** `print(df_avg_and_count_by_brand.columns.to_list())` should print `['brand', 'num_products', 'avg_price']`.
